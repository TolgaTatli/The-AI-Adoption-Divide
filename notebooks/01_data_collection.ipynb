## 6. Hızlı Görselleştirme

```python
# Top 10 ülke
top_10 = combined_df.nlargest(10, 'avg_interest')

plt.figure(figsize=(12, 6))
plt.barh(top_10['country_name'], top_10['avg_interest'])
plt.xlabel('Average Interest')
plt.title('Top 10 Countries by ChatGPT Interest')
plt.tight_layout()
plt.show()
```

```python
# GDP vs AI Interest
plt.figure(figsize=(12, 8))
plt.scatter(combined_df['gdp_per_capita'], combined_df['avg_interest'], 
           s=100, alpha=0.6)
plt.xlabel('GDP per Capita ($)')
plt.ylabel('AI Interest (Google Trends)')
plt.title('Economic Development vs AI Adoption')
plt.xscale('log')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
